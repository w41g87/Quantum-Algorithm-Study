```python
def getMatrix(a, b, c, d, e, f, g, h, j):
    return np.kron(np.kron(np.kron(np.kron(a, b), np.kron(c, d)), np.kron(np.kron(e, f), np.kron(g, h))), j)
```

```python
def getEig(m):
    w, v = np.linalg.eig(m)
    output = [v[:, j] for j in range(len(w)) if w[j] == 1]
    return output
```

```python
def intr(m, n):
    output = list()
    for w in m:
        for v in n:
            if np.array_equal(w, v):
                output.append(w)
    return output
```

```python
def checkM(m, e):
    output = list()
    for w in e:
        for v in e:
            output.append(eigTest(m, np.add(np.dot(1/math.sqrt(2), w), np.dot(1/math.sqrt(2), v))))
            output.append(eigTest(m, np.add(np.dot(-1/math.sqrt(2), w), np.dot(1/math.sqrt(2), v))))
            output.append(eigTest(m, np.add(np.dot(1/math.sqrt(2), w), np.dot(-1/math.sqrt(2), v))))
            output.append(eigTest(m, np.add(np.dot(-1/math.sqrt(2), w), np.dot(-1/math.sqrt(2), v))))

    return np.unique([x for x in output if x is not None], axis=0)
```

```python
def nullspace(A, atol=1e-13, rtol=0):
    A = np.atleast_2d(A)
    u, s, vh = sci.linalg.svd(A)
    tol = max(atol, rtol * s[0])
    nnz = (s >= tol).sum()
    ns = vh[nnz:].conj().T
    return ns
```
