```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
```

```python
secret = '10010110001100100'

circuit = QuantumCircuit(len(secret) + 1, len(secret))
for i in range(len(secret)):
    circuit.h(i)
    
circuit.x(len(secret))
circuit.h(len(secret))
circuit.barrier()

for i, n in enumerate(reversed(secret)):
    if n == '1':
        circuit.cx(i, len(secret))
        
circuit.barrier()
circuit.h(range(len(secret) + 1))
circuit.measure(range(len(secret)), range(len(secret)))
```

```python
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, simulator).result()
plot_histogram(result.get_counts(circuit))
```
